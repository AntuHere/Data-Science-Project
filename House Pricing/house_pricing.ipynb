```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('california_housing.csv')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

Drop the null value rows

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.columns
```

```python
sns.pairplot(df,  kind='scatter')
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='median_income', y='median_house_value', data=df, alpha=0.5)
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='longitude', y='latitude', data=df, alpha=0.5, hue='median_house_value', palette="husl")
```

```python
df_1 = df.drop(['ocean_proximity'], axis='columns')
df_1['Bedrooms Per Room'] = df_1['total_rooms'] / df_1['total_bedrooms']
df_1
```

```python
correlation_matrix = df_1.corr()
plt.figure(figsize=(15,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

Categorical Value

```python
dummy_values= pd.get_dummies(df['ocean_proximity']).astype(int)
dummy_values.head()
```

```python
dummy_values.value_counts()
```

```python
df_2 = pd.concat([df_1, dummy_values], axis=1)
```

```python
df_2
```

```python
df_2.shape
```

```python
correlation_matrix = df_2.corr()
plt.figure(figsize=(15,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```
